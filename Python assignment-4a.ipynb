1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in 
the list.

```python
def sum_of_even_numbers(numbers):
    # Using list comprehension and sum to calculate the sum of even numbers
    return sum(num for num in numbers if num % 2 == 0)

# Example usage
numbers = [1, 2, 3, 4, 5, 6]
result = sum_of_even_numbers(numbers)
print(result)
```

3.Implement a Python function that takes a list of integers and returns a new list containing the squares of 
each number.

```python
def square_numbers(numbers):
    return [num ** 2 for num in numbers]

# Example usage
numbers = [1, 2, 3, 4, 5]
squared_numbers = square_numbers(numbers)
print(squared_numbers) 
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, int(n ** 0.5) + 1):  # Check divisibility up to the square root of n
        if n % i == 0:
            return False
    return True

# Check prime numbers from 1 to 200
prime_numbers = [num for num in range(1, 201) if is_prime(num)]
print(prime_numbers)
```

5.Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of 
terms.

```python
class FibonacciIterator:
    def __init__(self, n):
        self.n = n  # Number of terms in the sequence
        self.a, self.b = 0, 1  # Starting values for Fibonacci sequence
        self.count = 0  # To keep track of how many terms we've generated

    def __iter__(self):
        return self  # The iterator object itself

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration  # Stop iteration if we've generated the required number of terms
        else:
            self.count += 1
            self.a, self.b = self.b, self.a + self.b  # Update the Fibonacci numbers
            return self.a  # Return the current Fibonacci number

# Example usage
fib = FibonacciIterator(10)  # Generate the first 10 Fibonacci numbers
for num in fib:
    print(num)
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent

```python
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

# Example usage
for power in powers_of_two(5):
    print(power)
```

7.  Implement a generator function that reads a file line by line and yields each line as a string

```python
def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()  # Use .strip() to remove any trailing newlines

# Example usage (ensure this is the correct file path)
file_path = r'C:\Users\Public\Documents\Hi Hello.txt'
for line in read_file_line_by_line(file_path):
    print(line)
```

8.  Use a lambda function in Python to sort a list of tuples based on the second element of each tuple

```python
# List of tuples
tuples_list = [(1, 3), (2, 1), (4, 2), (3, 4)]

# Sorting the list based on the second element of each tuple using a lambda function
sorted_list = sorted(tuples_list, key=lambda x: x[1])

# Output the sorted list
print(sorted_list)
```

9.  Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit

```python
# List of temperatures in Celsius
celsius_temps = [0, 25, 30, 100, -5]

# Convert Celsius to Fahrenheit using map and a lambda function
fahrenheit_temps = list(map(lambda celsius: (9/5) * celsius + 32, celsius_temps))

# Output the converted temperatures
print(fahrenheit_temps)
```

10.  Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
# Function to check if a character is not a vowel
def is_not_vowel(char):
    vowels = 'aeiouAEIOU'  # Define vowels (both lowercase and uppercase)
    return char not in vowels

# Input string
input_string = "ARIANA GRANDE"

# Use filter to remove vowels and join the remaining characters into a string
result_string = ''.join(filter(is_not_vowel, input_string))

# Output the result
print(result_string)
```
